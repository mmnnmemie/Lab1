#DataFrame Creation
Data are available in various forms and types like **CSV**, SQL table, JSON, or Python structures like **list**, **dict**, ndarrary, etc. We need to convert all such different data formats into a DataFrame so that we can use pandas libraries to analyze such data efficiently.

##DataFrame from list
**Lists** are used to store multiple items in a single variable. They are created using square brackets:

```python
import pandas as pd

# Create list
my_list = ['Apple', 10, 'Orange', 55.50]
print(my_list)
```

Topics

```python
import pandas as pd

# Create DataFrame from list
my_list = ['Apple', 10, 'Orange', 55.50]

my_list_df = pd.DataFrame(my_list)
print(my_list_df)
```

By default, all list elements are added as a row in the DataFrame.  Row index is the range of numbers(starting at 0). We can specify

*   row index into the index=[row_index1, row_index2] parameter in the DataFrame constructor
*   column labels into the columns=[col_labels] parameter in the DataFrame constructor.


```python
import pandas as pd

# Create list
fruits_list = ['Apple', 'Banana', 'Orange','Mango']

# Create DataFrame from list with a customized index
fruits_df = pd.DataFrame(fruits_list, index=['Fruit1', 'Fruit2', 'Fruit3', 'Fruit4'])
print(fruits_df)
```

```python
import pandas as pd

# Create list
fruits_list = ['Apple', 'Banana', 'Orange','Mango']

# Create DataFrame from list with a customized column name
fruits_df = pd.DataFrame(fruits_list, columns=['Fruits'])
print(fruits_df)
```

##DataFrame from dict
**dict** or dictionary is one Python data structure that is more generally known as an associative array. A dictionary consists of a collection of key-value pairs. Each key-value pair maps the key to its associated value.

You can define a dictionary by enclosing a comma-separated list of key-value pairs in curly braces ({}). A colon (:) separates each key from its associated value:

d = {
    key1: [value11, value12],
    key2: [value21, value21],
      .
      .
      .
    keyn: [valuen1, valuen2]
}

```python
import pandas as pd

# Python dict object
student_dict = {'Name': ['Joe', 'Nat'], 
                'Age': [20, 21], 
                'Marks': [85.10, 77.80]}
print(student_dict)
```

**Accessing Dictionary Values**

A value is retrieved from a dictionary by specifying its corresponding key in square brackets ([]):
